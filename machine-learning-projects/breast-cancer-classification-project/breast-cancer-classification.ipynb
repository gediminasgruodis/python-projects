```python
#Import modules
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
#Load breast cancer dataset
breast_cancer_data = load_breast_cancer()

#Split data into test and validation datasets
training_data, validation_data, training_labels, validation_labels  = train_test_split(breast_cancer_data.data, breast_cancer_data.target, test_size=0.2, random_state=100)

#Get the classifier scores and plot them on a graph
accuracies = []
k_list = [*range(1, 101)]

for k in range(1, 101):
    #Create a KNeighborsClassifier
    classifier = KNeighborsClassifier(n_neighbors = k)

    #Train the classifier
    classifier.fit(training_data, training_labels)

    #Get the classifier scores
    accuracy = classifier.score(validation_data, validation_labels)
    accuracies.append(accuracy)

#Visualize the classifier scores
plt.plot(k_list, accuracies)
plt.xlabel("k")
plt.ylabel("Validation accuracy")
plt.title("Breast Cancer Classifier Accuracy")
plt.show()
 

```
