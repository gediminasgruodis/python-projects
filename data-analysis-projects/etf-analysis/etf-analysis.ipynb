```python
import pandas as pd
import numpy as np
import yfinance as ytf
plt.style.use(style='seaborn')
%matplotlib inline
```

```python
vwce = ytf.Ticker('VWCE.DE')
vwce = vwce.history(period='max')
vwce['Ticker'] = 'VWCE.DE'

iwda = ytf.Ticker('IWDA.AS')
iwda = iwda.history(period='max')
iwda['Ticker'] = 'IWDA.AS'

sxr8 = ytf.Ticker('SXR8.DE')
sxr8 = sxr8.history(period='max')
sxr8['Ticker'] = 'SXR8.DE'

iusn = ytf.Ticker('IUSN.DE')
iusn = iusn.history(period='max')
iusn['Ticker'] = 'IUSN.DE'

dataset = pd.concat([vwce, iwda, sxr8, iusn])
dataset.head()
```

```python
# Plot all the close prices
vwce['Close'].plot(figsize=(10, 7))
iwda['Close'].plot(figsize=(10, 7))
sxr8['Close'].plot(figsize=(10, 7))
iusn['Close'].plot(figsize=(10, 7))


# Show the legend
plt.legend(['VWCE', 'IWDA', 'SXR8', 'IUSN'])

# Define the label for the title of the figure
plt.title("Close Price", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="minor", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
